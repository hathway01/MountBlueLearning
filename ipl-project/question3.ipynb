```python
import matplotlib.pyplot as plt

import csv

with open('ipl/matches.csv') as csvfile:
    reader1 = csv.reader(csvfile,delimiter=",")
    next(reader1)
    
    idlist=[]
    for row in reader1:
        if(int(row[1])==2016):
            idlist.append(row[0])
csvfile.close()
    
result={}
    
with open('ipl/deliveries.csv') as csvfile:
    reader2 = csv.reader(csvfile,delimiter=",")
    next(reader2)
    for row in reader2:
        if(row[0] in idlist):
            if row[3] not in result:
                result[row[3]]=int(row[16])
            else:
                result[row[3]]+=int(row[16])
csvfile.close()

for team in result:                
    plt.bar(team,result[team])

plt.style.use('ggplot')
plt.title("Runs Conceded By Teams")
plt.ylabel("Runs")
plt.xlabel("Teams")
plt.xticks([team for team in range(len(result.keys()))],result,rotation=90)
plt.show()
   
            
```
